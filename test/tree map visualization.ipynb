```python
group= base.groupby(["StateDesc","CityName"])
meanvalue = group["Data_Value"].mean()
```

```python
USDA = pd.read_csv('../data/USDA_Farmers_Market.csv')
USDA.head()
```

```python
merged_df = USDA.merge(meanvalue, how = 'left', on = ['CityName', 'StateDesc'])
merged_df
```

```python
merged_df=merged_df.dropna(subset=["Data_Value"])
merged_df
```

```python
fig1, ax1 = plt.subplots()
ax1.boxplot(merged_df["Data_Value"])
```

```python
c_merged_df= merged_df.drop(columns=["FMID","Website","Facebook","Twitter","Youtube","OtherMedia","street","updateTime",
                                   "Season1Date","Season1Time","Season2Date","Season2Time","Season3Date","Season3Time",
                                    "Season4Date","Season4Time",'Location', 'Credit', 'WIC', 'WICcash'])
```

```python
c_merged_df.columns
```

```python
top = c_merged_df.loc[(c_merged_df["Data_Value"]>=38.289772998805226)]
low = c_merged_df.loc[(c_merged_df["Data_Value"]<=22.525)]
```

```python
top["Organic"].value_counts()
```

```python
import squarify
```

```python
#Dataframe includes City, State, Data Value and total Number of Farmers Markets for top 10% of cities
C_top = pd.DataFrame(top.groupby(["CityName","StateDesc", "Data_Value"]).size())
C_top = C_top.reset_index()
C_top = C_top.rename(columns={0:"total"})
C_top

labels = C_top.apply(lambda x: str(x[0]) + "\n (" + str(x[1]) + ")", axis=1)
sizes = C_top['Data_Value'].values.tolist()
colors = [plt.cm.Spectral(i/float(len(labels))) for i in range(len(labels))]

# Draw Plot
plt.figure(figsize=(28,19), dpi= 80)
squarify.plot(sizes=sizes, label=labels, color=colors, alpha=.5)

# Decorate
plt.title('Number of Farmers Markets in Cities within top 10% Obesity Value', fontsize=23,fontweight="bold")
plt.axis('off')
plt.show()
```

```python
C_top
```

```python
C_low = pd.DataFrame(low.groupby(["CityName","StateDesc", "Data_Value"]).size())
C_low = C_low.reset_index()
C_low = C_low.rename(columns={0:"total"})
#C_low
labels = C_low.apply(lambda x: str(x[0]) + "\n (" + str(x[1]) + ")", axis=1)
sizes = C_low['total'].values.tolist()
colors = [plt.cm.Spectral(i/float(len(labels))) for i in range(len(labels))]

# Draw Plot
plt.figure(figsize=(40,20), dpi= 80)
squarify.plot(sizes=sizes, label=labels, color=colors, alpha=.8)

# Decorate
plt.title('Number of Farmers Markets in Cities within bottom 10% Obesity Value',fontsize=23,fontweight="bold")
plt.axis('off')
plt.show()
```
